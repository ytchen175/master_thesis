```python
# stanford_alpaca 的 prompt template
# ref: https://zhuanlan.zhihu.com/p/647149346
PROMPT_DICT = {
    "prompt_input": (
        "Below is an instruction that describes a task, paired with an input that provides further context. "
        "Write a response that appropriately completes the request.\n\n"
        "### Instruction:\n{}\n\n### Input:\n{}\n\n### Response:{}"
    ),
    "prompt_no_input": (
        "Below is an instruction that describes a task. "
        "Write a response that appropriately completes the request.\n\n"
        "### Instruction:\n{}\n\n### Response:{}"
    ),
}
```

```python
df['text'] = [np.nan] * len(df)
```

```python
for idx, row in df.iterrows():
    # example = row[1]
    example = row
    output_texts = []

    if type(example["input"]) != np.nan:
        text = PROMPT_DICT['prompt_input'].format(example["instruction"], example["input"], example["output"])
    else:
        text = PROMPT_DICT['prompt_no_input'].format(example["instruction"], example["output"])

    # output_texts.append(text)
    df.at[idx, 'text'] = text
```
