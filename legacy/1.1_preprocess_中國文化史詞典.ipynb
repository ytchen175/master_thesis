```python
def select_only_chinese_characters(inputString: str):
    return "".join(re.findall(r'[\u4E00-\u9FFF]', inputString)) # chinese regex

def clean_all_html_tags(raw_text: str):
    soup = BeautifulSoup(raw_text)
    p_text = soup.get_text().strip()
    return p_text
```

```python
df['key'] = df['key'].apply(lambda x: select_only_chinese_characters(x))
df['value'] = df['value'].apply(lambda x: clean_all_html_tags(x))
```

```python
# 可以發現此語料品質不佳，許多是機翻(eg. 2604, 2924, 4152)，因此剔除掉非對稱簡繁字雖然粗暴，但保證了正體語料的 100% 正確性
not_symmetric_characters_in_SC_to_TC_df = pd.read_csv('./corpus/非對稱簡繁字/簡化字對照標準字總表.csv')

not_symmetric_characters = []
not_symmetric = not_symmetric_characters_in_SC_to_TC_df[not_symmetric_characters_in_SC_to_TC_df['非對稱簡繁字']]

for s in not_symmetric['正體字'].tolist():
    l = eval(s)  # NOTE: the input is a fixed file, so I use `eval`
    not_symmetric_characters.extend(l)
```

```python
not_symmetric_characters_regex = "[" + '|'.join(not_symmetric_characters) + "]"

# 剔除掉含非對稱簡繁字 & 要參考到別的詞條的 rows
df = df[~df['value'].str.contains(not_symmetric_characters_regex) & ~df['value'].str.contains('見“(.*)”')]
```

```python
t2s_converter = opencc.OpenCC('t2s.json')

"""traditional chinese to simplified chinese"""
def convert_TC_to_SC(zhtw_text: str, converter=None):
    if not converter:
        converter = opencc.OpenCC('t2s.json')
    return converter.convert(zhtw_text)
```

```python
# Start making instruct dataset, columns follows Alpaca dataset format: instruction, input, output, and category for classify tasks

SEED_STUB = 77777
count = 1

instruction_dataset = pd.DataFrame()

category_lst = []
instruction_lst = []
input_lst = []
output_lst = []

random.seed(SEED_STUB)

for i in df['value']:
    for lan in translate_task_instructions.keys():
        instruction_pool = list(translate_task_instructions[lan].values())

        category_lst.append("translate_from_SC")
        instruction_lst.append(random.choice(instruction_pool))
        input_lst.append(convert_TC_to_SC(i, t2s_converter))
        output_lst.append(i)

        count += 1
        random.seed(SEED_STUB + count)

instruction_dataset['category'] = category_lst
instruction_dataset['instruction'] = instruction_lst
instruction_dataset['input'] = input_lst
instruction_dataset['output'] = output_lst
```

```python
category_lst = []
instruction_lst = []
input_lst = []
output_lst = []

count = 1

interpret_instruction_dataset = pd.DataFrame()

random.seed(SEED_STUB)

for index, row in df.iterrows():
    k, v = row
    for lan in interpret_task_instructions.keys():
        instruction_pool = list(interpret_task_instructions[lan].values())

        prompt = random.choice(instruction_pool + [common_prompt])

        category_lst.append("interpret")

        if "{}" in prompt:
            instruction_lst.append(prompt.format(k))
            input_lst.append("")
        else:
            instruction_lst.append(prompt)
            input_lst.append(k)

        output_lst.append(v)

        count += 1
        random.seed(SEED_STUB + count)

interpret_instruction_dataset['category'] = category_lst
interpret_instruction_dataset['instruction'] = instruction_lst
interpret_instruction_dataset['input'] = input_lst
interpret_instruction_dataset['output'] = output_lst
```
