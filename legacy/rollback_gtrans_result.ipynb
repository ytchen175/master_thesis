```python
import time
import csv
import json
import pandas as pd
import googletrans

from pprint import pprint
from tqdm.auto import tqdm

import opencc

t2s_converter = opencc.OpenCC('t2s.json')

"""traditional chinese to simplified chinese"""
def convert_TC_to_SC(zhtw_text: str, converter=None):
    if not converter:
        converter = opencc.OpenCC('t2s.json')
    return converter.convert(zhtw_text)

translator = googletrans.Translator()

filename = 'google_translate_result.csv'
df = pd.read_csv('test_dataset.csv')

# sentences = df['ch_SC'].to_list()

# 回補資料
res = pd.read_csv('google_translate_result.csv')
ch_SC_google = list(map(convert_TC_to_SC, res['ch_google']))

```

```python
sentences = set(df['ch_SC'].to_list()) - set(ch_SC_google)

print(len(df['ch_SC'].to_list()))
print(len(ch_SC_google))
print(len(set(df['ch_SC'].to_list()) & set(ch_SC_google)))
```

```python
res['ch_SC_google'] = ch_SC_google
```
